Linear Regression case 1

```python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
model = LinearRegression()
```

```python
X = np.array([
    [1], [1.5], [2], [2.2], [2.5], 
    [2.8], [3], [3.2], [3.5], [3.7],
    [4], [4.3], [4.5], [4.7], [4.8],
    [5], [5.2], [5.5], [5.7], [5.8],
    [6], [6.3], [6.5], [6.7], [6.8],
    [7], [7.2], [7.5], [7.8], [8]
])

y = np.array([
    28000, 32000, 35000, 36000, 37000,
    38500, 40000, 41000, 42500, 43500,
    45000, 46500, 47500, 48500, 49000,
    50000, 51000, 52500, 53500, 54000,
    55000, 56500, 57500, 58500, 59000,
    60000, 61000, 62500, 63500, 65000
])
```

```python
case1_model = model.fit(X, y)
Predicted_salary = case1_model.predict([[6.5]])
print(f"Predicted salary for 6.5 years of experience: {Predicted_salary[0]}")
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Experience (years)')
plt.ylabel('Salary')
plt.title('One-variable Regression (y is scalar)')
plt.show()
```
